# 02 - Create Lagged GRACE-Only Forecasting Dataset

Features are past GRACE TWSA only. The target is TWSA at the current forecast month.

```python
from pathlib import Path
import sys
# Resolve local package imports from common notebook launch locations.
ROOT = Path.cwd() if (Path.cwd() / 'src').exists() else Path.cwd().parent
sys.path.append(str(ROOT / 'src'))

from grace_gnn.config import BASIN_MONTH_CSV, LAGGED_DATASET_CSV, LAGS, ensure_dirs
from grace_gnn.data import read_basin_month_csv
from grace_gnn.features import make_lagged_dataset

# Load cleaned basin-month observations for lag feature construction.
ensure_dirs()
basin_month = read_basin_month_csv(BASIN_MONTH_CSV)
```

```python
# Build autoregressive lag features only after the cleaned table exists.
if basin_month.empty:
    print('No basin-month data available. Run notebook 01 after placing the required data files.')
    lagged = None
else:
    # Persist the lagged modeling table for downstream training notebooks.
    lagged = make_lagged_dataset(basin_month, LAGS, LAGGED_DATASET_CSV)
    print(f'Saved lagged dataset with {len(lagged):,} rows to {LAGGED_DATASET_CSV}')
    print('Feature lags:', LAGS)
    display(lagged.head())
```
